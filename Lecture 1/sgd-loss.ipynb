We want to find the line passing through the origin $y = mx$ which minimize the distance from each point (the error)

The distance between a point $(x_0,y_0)$ and a line $ax + by + c = 0$ is:

$$
E(a,b,c,x_0,y_0)=\frac{|ax_0 + by_0 + c|}{\sqrt{a^2 + b^2}}
$$

We can rewrite it for our use-case.

Our model (the line passing through the origin) is: $y = mx$.

So $ax + y = 0$ and then $y = -ax$ which means $a = -m$

The distance between a point $(x_0,y_0)$ and the line $y = mx$ is then (with $a=-m$, $b=1$, $c=0$)

$$
E(m,x_0,y_0)=\frac{|-mx_0 + y_0|}{\sqrt{m^2 + 1}}
$$

```python
def error(x, y, m):
    return (np.abs(-m*x + y) / np.sqrt(m**2 + 1)).mean()
```

Lets plot the error with respect to some values of $m$ ranging from $0$ to $10$

```python
some_m = np.linspace(0, 10, 1000)
some_error = [np.sum(error(x, y, i)) for i in some_m]

plt.figure()
plt.xlabel('m')
plt.ylabel('error')
plt.plot(some_m, some_error)
plt.show()
```

We can compute the derivative $\frac{\partial E}{\partial m}$ of $E$ with respect of $m$.

$$
\frac{\partial E(m, x_0, y_0)}{\partial m} = \frac{(m^2 + 1)x_0(mx_0 - y_0) - m(y_0-mx_0)^2}{(m^2 + 1)^\frac{3}{2}|y_0-mx_0|}
$$

```python
def error_grad(x, y, m):
    return (((m**2 + 1)*x*(m*x - y) - m*(y - m*x)**2)/((m**2 + 1)**(3/2) * np.abs(y-m*x))).mean()
```

Starting from a random $m$

We can:
- Compute $\frac{\partial E}{\partial m}$ (which points in the direction that maximize $E$)
- Update $m$ in the direction $-\frac{\partial E}{\partial m}$ (direction that minimize $E$)
- Loop for some iterations

```python
m = 0.1

m_hist = []
e_hist = []
for i in range(0, 2_000):
    m = m - 0.001 * error_grad(x, y, m)
    e = error(x, y, m)

    e_hist.append(e)
    m_hist.append(m)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(m_hist)
ax1.set_xlabel('iteration')
ax1.set_ylabel('m')

ax2.plot(e_hist)
ax2.set_xlabel('iteration')
ax2.set_ylabel('error')

plt.show()
```

This is called **Gradient Descent**

If instead of computing the error $E$ over the whole dataset we just use a random subset

```python
m = 0.1

m_hist = []
e_hist = []
for i in range(0, 2_000):
    x_batch = np.random.choice(x, 100)
    y_batch = np.random.choice(y, 100)
    
    m = m - 0.001 * error_grad(x_batch, y_batch, m)
    e = error(x, y, m)

    e_hist.append(e)
    m_hist.append(m)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(m_hist)
ax1.set_xlabel('iteration')
ax1.set_ylabel('m')

ax2.plot(e_hist)
ax2.set_xlabel('iteration')
ax2.set_ylabel('error')

plt.show()
```

It is called **Stochastic Gradient Descent**

Lets put it all togheter

```python
from PIL import Image

some_m = np.linspace(0, 10, 1000)
some_error = [np.sum(error(x, y, i)) for i in some_m]
xx = np.linspace(-25, 25, 1000)

m = 0.1
frames = []
for i in range(0, 2_000):
    grad = error_grad(x, y, m)
    err = error(x, y, m)

    if i % 50 == 0:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

        ax1.plot(some_m, some_error)
        ax1.scatter(m, err, color='red')

        ax2.scatter(x, y, s=4)
        ax2.plot(xx, m*xx, color='red')

        ax2.set_xlim(-25, 25)
        ax2.set_ylim(-25, 25)
        ax2.axhline(0, color='black', linewidth=.5)
        ax2.axvline(0, color='black', linewidth=.5)
        fig.canvas.draw()
        frames.append(Image.frombytes('RGB', fig.canvas.get_width_height(),fig.canvas.tostring_rgb()))

        plt.close()
    
    m = m - 0.001 * grad
```
